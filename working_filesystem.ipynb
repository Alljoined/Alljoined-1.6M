```python
from pathlib import Path

# Set base directory
base_dir = Path("/srv/eeg_reconstruction/shared/data/preprocessed_data/Alljoined-1.7M")

# Collect all sub-XX directories (sub-01 to sub-20)
subject_dirs = sorted([d for d in base_dir.iterdir() if d.is_dir() and d.name.startswith("sub-")])

# Tracking
subjects_with_files = {}
subjects_without_files = []

# Loop through subjects
for subj_dir in subject_dirs:
    files = list(subj_dir.glob("*"))  # all files in subject folder
    
    if files:
        subjects_with_files[subj_dir.name] = [f.name for f in files]
    else:
        subjects_without_files.append(subj_dir.name)

# Print results
print("✅ Subjects with files:")
for subj, files in subjects_with_files.items():
    print(f"  {subj}: {len(files)} file(s)")
    for f in files:
        print(f"    - {f}")

print("\n❌ Subjects without any files:")
for subj in subjects_without_files:
    print(f"  {subj}")
```

```python
target_suffix = "h_freq_25.0_RJ_None.npy"

# Collect all sub-XX directories
subject_dirs = sorted([d for d in base_dir.iterdir() if d.is_dir() and d.name.startswith("sub-")])

# Track subjects missing the suffix
subjects_missing_suffix = []

# Loop through subjects
for subj_dir in subject_dirs:
    matching_files = list(subj_dir.glob(f"*{target_suffix}"))
    
    if not matching_files:
        subjects_missing_suffix.append(subj_dir.name)

# Output results
print(f"❌ Subjects WITHOUT any files ending in '{target_suffix}':")
for subj in subjects_missing_suffix:
    print(f"  {subj}")
```

```python
import pyarrow.parquet as pq

path = "/srv/eeg_reconstruction/shared/data/preprocessed_data/repr_eeg2/sub-04/experiment_metadata.parquet"
path2 = "/srv/eeg_reconstruction/shared/data/preprocessed_data/Alljoined-1.7M/sub-19/experiment_metadata.parquet"

schema1 = pq.read_schema(path)
schema2 = pq.read_schema(path2)

print(schema1.equals(schema2))  # True if identical
print(schema1)                  # See details

print(schema2)
```

```python
import pandas as pd

df1 = pd.read_parquet(path)
df2 = pd.read_parquet(path2)

# Check shape
print(df1.shape == df2.shape)

# Sort and compare
equal = df1.sort_values(by=df1.columns.tolist()).reset_index(drop=True).equals(
         df2.sort_values(by=df2.columns.tolist()).reset_index(drop=True))
print("Equal:", equal)
```
